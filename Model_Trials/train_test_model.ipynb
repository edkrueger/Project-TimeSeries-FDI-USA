```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline

from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv("../Resources/full_data.csv").set_index("countries")
```

```python
df
```

```python
target = "fdi_in_usa_million"
```

```python
X = df.drop(
    [
        "fdi_in_usa_million",
        "year_id",
        "country_id",
        "regions",
        "region_id",
        "western_emerging",
        "west_emerge_id",
        "capital",
        "latitude",
        "longitude",
        "interaction_gdp",
        "interaction_us_fdi",
        "interaction_stock_mkt",
    ],
    axis=1,
).values
y = df[target].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
y_test_pred = linear_regression.predict(X_test)
mean_squared_error(y_test, y_test_pred)
```

```python
decsion_tree = DecisionTreeRegressor()
decsion_tree.fit(X_train, y_train)
y_test_pred = decsion_tree.predict(X_test)
mean_squared_error(y_test, y_test_pred)
```

```python
random_forest = RandomForestRegressor()
random_forest.fit(X_train, y_train)
y_test_pred = random_forest.predict(X_test)
mean_squared_error(y_test, y_test_pred)
```
